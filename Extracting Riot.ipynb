### Function for getting the accountId of a summoner

```python
def accId(summonerName, APIKey):
    # Puts the input of both the APIKey and the summonerName into a URL string
    URL = "https://na1.api.riotgames.com/tft/summoner/v1/summoners/by-name/" + summonerName + "?api_key=" + APIKey
    # Requests the information from the given URL
    response = requests.get(URL)
    # Turns the response into json format for 
    r1 = response.json()
    print(json.dumps(r1, indent = 3, sort_keys=True))
    # Returns the accountId
    return r1['accountId']
```
